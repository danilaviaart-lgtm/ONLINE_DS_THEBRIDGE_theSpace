## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1

com1 = A and C == B
print(com1)
com2 = A != C
print(com2)
com3 = A > C
print(com3)
com4 = C <= A
print(com4)
com4 = B != C 
print(com4)

````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
pregunta1 = input("Escribe un número")
pregunta2 = input("Escribe otro número")

igualdad = pregunta1 == pregunta2 ## para probar maneras de hacerlo
print(bool(igualdad)) ## para probar

if bool(igualdad) == True:
    print("Los números son iguales")
else:
    print("Los números no son iguales")
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
pregunta1 = input("Escribe un número")
pregunta2 = input("Escribe otro número")
pregunta3 = input("Escribe otro númer más")

if pregunta1 == pregunta2 or pregunta1 == pregunta3 or pregunta2 == pregunta3:
    print("Hay números que coinciden.")
else:
    print("No hay números iguales.")
````
